```python

df = pd.get_dummies(df, drop_first=True)
df
```

```python
sum = []
for val in df.select_dtypes(include=['int64','float64']).columns:
    q1 = df[val].quantile(0.25)
    q3 = df[val].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    out = df[(df[val] < lower) | (df[val] > upper)]
    sum.append((val, len(out)))
print(sum)
```

```python
x=df.drop(['fraud_reported_Y','umbrella_limit'],axis=1)
y=df['fraud_reported_Y']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
df=ms.fit_transform(df)
df
```

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier()
dc.fit(x_train,y_train)
```

```python
y_pred=dc.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("accuracy score:",accuracy_score(y_test,y_pred))
print("confusion matrix:",confusion_matrix(y_test,y_pred))
print("classification report:",classification_report(y_test,y_pred))
```
